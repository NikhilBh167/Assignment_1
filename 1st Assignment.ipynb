```python
# 1. Import the data set into a Pandas DataFrame.

# Import panda and numpy

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
movies_df=pd.read_csv(r'C:\Users\Nikhil\Downloads\Movie Assignment Data.csv')
movies_df
```

```python
# 2. Generate descriptive statistics for the budget of all the movies.

#Using describe function to get all the statistics values of movie data

movies_df.describe()
```

```python
# 3. Find out how many of the top-rated movies produced in the United States have a PG-13 rating.


# Using '==' function to get top-rated movies produced in the United States have a PG-13 rating.

movies_df[(movies_df['Country']=='USA') & (movies_df['content_rating']=='PG-13')].shape[0]
```

```python
# 4. Find out whether any of the top-rated movies produced in 2014 were not produced in the United States.

# Using '==' & '!=' function to get top-rated movies produced in 2014 were not produced in the United States.

movies_df[(movies_df['title_year'] == 2014) & (movies_df['Country']!='USA')].shape[0]
```

```python
# 5. Find the percentage of the top-rated movies that are in: 1 genre only

# Using round function and count function

round(movies_df.genre_1.count()-movies_df.genre_2.count()*100/movies_df.Title.count(),2)
```

```python
# 5. Find the percentage of the top-rated movies that are in: 2 genre only

# Using round function and count function

round(movies_df.genre_2.count()-movies_df.genre_3.count()*100/movies_df.Title.count(),2)
```

```python
# 5. Find the percentage of the top-rated movies that are in: 3 genre only

# Using round function and count function


round(movies_df.genre_3.count()*100/movies_df.Title.count(),2)
```

```python
# 6. Convert the budget and gross values from “dollars” to “dollars in millions” for all top-rated movies. Round the converted values down to 3 decimal places. For example, a value of 192,345,273 should be converted to 192.345.

# For converting dollar to millions dollar divide budget and gross with 1000000

movies_df['budget']=np.round(movies_df['budget']/1000000, 3)
movies_df['Gross']=np.round(movies_df['Gross']/1000000, 3)
movies_df


```

```python
# 7. List all details for the top 10 movies with the highest profit, sorted from highest to lowest. Hint: Profit is not a column in the DataFrame. You will need to calculate it.

#Make profit column 
movies_df['Profit']=movies_df['Gross']-movies_df['budget']

# Using sort_value function by Profit
movies_df=movies_df.sort_values(by='Profit',ascending=False).head(10)
movies_df
```

```python
# 8. Generate a list of all the actors, in alphabetical order by the first name, 
#that have starred in a top-rated movie. If an actor has starred in multiple movies, 
#their name should appear only once on the list. Assume that all actors’ names are in the format <first_name> <last_name>.


# Create series of actor 1,2 & 3

actor_1= movies_df['actor_1_name']
actor_2= movies_df['actor_2_name']
actor_3= movies_df['actor_3_name']

# Using 'append ' function to add actor 1,2 & 3

all_actors = actor_1.append(actor_2).append(actor_3).sort_values()
all_actors

# For removing zero we use '.remove()' function

all_actors_list = list(all_actors.unique())
all_actors_list.remove('0')
all_actors_list

```

```python
all_actors.value_counts().head(10)
```

```python
# 9. Assuming you have a DataFrame named top_rated_movies

# Assuming 'actor_1', 'actor_2', and 'actor_3' are the columns representing the actors
# Combine these columns into a single list of all actors
actors = movies_df['actor_1_name'].tolist() + movies_df['actor_2_name'].tolist() + movies_df['actor_3_name'].tolist()

# Create a Pandas Series to count the occurrences of each actor
actor_counts = pd.Series(actors).value_counts()

# Get the top 3 actors with the highest counts
top_3_actors = actor_counts.head(3)

top_3_actors
```

```python
# 10. Create a data visualization that shows each country and 
# the number of top-rated movies produced in it. Find the country that produced the most top-rated movies.


# First we define country & title and using coount & sort function

movies_status_df= movies_df.groupby(['Country']).count()[['Title']].sort_values(by='Title',ascending= False)
movies_status_df
```

```python
# After that we are using plot function for ploting bar graph

# Using plot function for bar graph and mention x & Y lable for the graph

movies_status_df.plot(kind='bar')
plt.title('Number of Top-rated movies by country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```
